```python
import math
import time
import numpy as np
import pandas as pd
import datetime as dt
import cufflinks as cf
from pylab import plt
```

```python
np.random.seed(1000)
plt.style.use("seaborn")
%matplotlib inline
```

```python
p = 0.55
```

```python
f = p - (1-p)
```

```python
f
```

```python
I = 50
```

```python
n = 100
```

```python
 def run_simulation(f): 
        c = np.zeros((n, I))
        c[0] = 100
        for i in range(I):
            for t in range(1, n):
                o = np.random.binomial(1, p) 
                if o > 0:
                    c[t,i]=(1+f)*c[t-1,i] 
                else:
                    c[t,i]=(1-f)*c[t-1,i]
        return c
```

```python
c_1 = run_simulation(f)
```

```python
c_1.round(2)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(c_1, 'b', lw=0.5)
plt.plot(c_1.mean(axis=1), 'r', lw=2.5);
```
